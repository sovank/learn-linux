Now, after installing Apache, it's time to run Apache under a dedicated user named apache_admin to enhance security. Operating Apache under its own non-privileged user account isolates the processes from other system activities, which is a common security practice.

Create a dedicated user named apache_admin with no login shell access.

```python
# Answer
# sudo useradd -r -s /usr/sbin/nologin apache_admin
# -r is for creating a system account
# System accounts are typically used for running services,
# and they generally have no home directories and no login capabilities
# -s is for shell
```

# Question 13

TASK - 1:

You need to adjust the configuration of the Apache web server to run under the newly created apache_admin user account. This ensures that all Apache processes use this non-privileged user, improving the server's security. Open the Apache main configuration file in a text editor with root privileges. It is located at /etc/apache2/apache2.conf. Make sure to restart the Apache after making changes in the configuration file.

```python
# Answer
# sudo vi /etc/apache2/apache2.conf

# In the configuration file, find and update the User and Group directives to apache_admin Find these lines:
# User ${APACHE_RUN_USER}
# Group ${APACHE_RUN_GROUP}

# And Replace them with:

# User apache_admin
# Group apache_admin
```



---



TASK - 2:

To secure the web directory and maintain operational integrity, change the ownership of the /var/www/html directory to the apache_admin user and group. This step is crucial for preventing unauthorized access to and modifications to the web server's files. After all the above steps, make sure to restart the Apache server.

```python
# Answer
# sudo chown -R apache_admin:apache_admin /var/www/html
```



---



TASK - 3:

Enable the apache_admin user to restart the Apache web server without needing to enter a password, simplifying server management while maintaining security. Edit the sudoers file to attempt this task.

Verify the configuration after updating the sudoers file to ensure it works as intended. Switch to the apache_admin user and attempt to restart the Apache service.


```python
# sudo vi /etc/sudoers.d/apache_admin
# sudo visudo -f /etc/sudoers.d/apache_admin
# apache_admin ALL=(ALL) NOPASSWD: /bin/systemctl restart apache2
# sudo su - apache_admin
# sudo systemctl restart apache2
```

commands.

- Viewing the contents of the /var/log/alternatives.log file

- Restarting the nginx service

- Checking the status of the nginx service
