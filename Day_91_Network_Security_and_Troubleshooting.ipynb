# UFW Firewall Configuration Tasks

image.png



---



**Task 1:**


Turn On UFW
Description: Activate the firewall to start protecting the server.

```python
sudo ufw enable
```



---



**Task 2:**

Allow SSH (Port 22)
Description: Open port 22 for SSH to enable remote server management.

```python
sudo ufw allow 22
```



---



**Task 3:**

Allow HTTP Traffic (Port 80)
Description: Open port 80 to let web browsers access the server.

```python
sudo ufw allow 80
```



---



**Task 4:**

Allow DNS Requests (Port 53, TCP)
Description: Open port 53 through TCP protocol to support handling large DNS queries.

```python
sudo ufw allow 53/tcp
```



---



**Task 5:**

Block HTTPS Traffic (Port 443)


Description:
You need to stop all secure web traffic by blocking port 443 through TCP Protocol.

```python
sudo ufw deny 443/tcp
```
