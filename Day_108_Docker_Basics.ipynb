# Q2. Creating a Custom Docker Image

image.png

```python
cd /home/user/student/assignments

sudo vi Dockerfile

FROM python:3.8

WORKDIR /app

COPY . /app # dot (.) here means everything from current

CMD ["python", "app.py"]
```

image.png

```python
docker build -t my-python-app .  # (.) represents the current directory
```
